```python
# 問題提供サイトで汎用的に使用されている形式の問題文テキストを解析し、配列へ変換します

def parse_numberlink_data(raw_data: str):
    """
    ドット(.)と数字で構成された文字列データを
    ソルバー用の2次元配列(0と整数)に変換する
    """
    grid = []
    
    # 行ごとに分割し、空行を除外して処理
    lines = raw_data.strip().splitlines()
    
    for line in lines:
        # 空白文字(スペースやタブ)で分割
        tokens = line.split()
        if not tokens:
            continue
            
        row_data = []
        for token in tokens:
            if token == '.':
                # 空きマスは 0 に変換
                row_data.append(0)
            else:
                try:
                    # 数字は整数に変換
                    row_data.append(int(token))
                except ValueError:
                    # 万が一数字以外(例: アルファベット)が混ざっていた場合の対策
                    # 必要に応じてここを拡張（A->1, B->2など）できます
                    print(f"Warning: Unexpected token '{token}' found. Treating as 0.")
                    row_data.append(0)
        
        grid.append(row_data)
            
    return grid

# --- 実行テスト ---

input_text = """
5 . . . . . . . 7 . . . 
. . . . . . 4 . . 5 1 . 
. . . . . . . . . 1 . . 
. . . . . . . 9 . . 6 . 
. . . . . . . . . . 8 . 
. . . . . . . 11 3 . . . 
. 10 . . . 3 . . . . . . 
. . . . . . . . 11 . . . 
. . 6 . . . . . 10 . . . 
. . . . . . . . . . . . 
. . 7 . . . 8 . . . . . 
4 . . . 9 . . . . . . . 
"""

# パース実行
parsed_grid = parse_numberlink_data(input_text)

# 結果の確認
print("--- Parsed Grid for Solver ---")
for row in parsed_grid:
    print(row)
```
