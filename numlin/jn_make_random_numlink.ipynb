```python
import arukone_gen
import SaveNumlinkFormat
import numpy as np
import time
from datetime import datetime
from tqdm import tqdm

data = []

h, w = 10, 10
print(f"--- {h}x{w} 全マス使用ナンバーリンク作問中 ---")

for i in tqdm(range(10000)):
    t0 = time.time()
    grid, paths, conf = arukone_gen.build_arukone(
        h, 
        w, 
        target_max_n=5, 
        min_conflicts=0, 
        verbose=False
        )
    url = SaveNumlinkFormat.generate_numlin_url(w, h, [i for r in grid for i in r])
    elapsed = time.time() - t0

    data.append([h, w, len(paths), conf, elapsed, url])



print(f"★ 終了  ({elapsed:.2f}秒)")
```

```python
import pandas as pd

pd.DataFrame(data, columns=["h", "w", "nums", "conf", "elapsed", "url"]).to_pickle("numlink_problems_data_master.pkl")
```
